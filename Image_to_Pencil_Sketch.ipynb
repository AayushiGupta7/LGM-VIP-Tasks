## Project : Image to Pencil Sketch using Python

We need to read image in RGB format and then convert it to a grey scale image.This will turn an image  into a classic black and white photo.Then the next thing to do is invert the gray scale image also called negative image, this will be our inverted image gray scale image.Inversion can be used to enhance details. Then we can finally create pencil sketch by mixing the gray scale image with the inverted blurry image.This can be done by dividing the gray scale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

### Importing necessary libraries

```python
import cv2 
from PIL import Image
```

### Reading Image

```python
image = cv2.imread('butterfly.jpg')
```

### Converting the image to RGB format

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Displaying the Image

```python
display(Image.fromarray(image))
```

### Converting the image to gray scale image

```python
grey_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

### Dislaying the gray scale image

```python
display(Image.fromarray(grey_image))
```

### Inverting the gray scale image

```python
inverted_grey_image = 255-grey_image
```

### Displaying the inverted gray scale image(negative image

```python
display(Image.fromarray(inverted_grey_image))
```

### Blurring the negative image

```python
blurry_negative_image = cv2.GaussianBlur(inverted_grey_image, (21,21),0)
```

### Displaying the blurry negative image

```python
display(Image.fromarray(blurry_negative_image))
```

### Inverting the  blurry  neagtive image

```python
inverted_blurry_image = 225-blurry_negative_image
```

### Displaying the inverted blurry image

```python
display(Image.fromarray(inverted_blurry_image))
```

### Dividing the gray scale image by the inverted blurry image

```python
divided_image = cv2.divide(grey_image,inverted_blurry_image,scale = 256.0)
```

### Displaying the divided image

```python
display(Image.fromarray(divided_image))
```

### Comapring both the images

### Displaying both the images

```python
display(Image.fromarray(image))
display(Image.fromarray(divided_image))
```
